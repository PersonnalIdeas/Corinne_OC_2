```python
# 2.1.1 Traitement des colonnes avec un taux de valeurs manquantes supérieures 65%

# 2.1.1 Première option: utiliser résultats du data profiling

#Suppression des colonnes identifiées comme ayant un taux de remplissage < 35%
#liste des colonnes à suprimer non complète juste pour montrer la commande python 
df.drop(['alcohol_100g', 'allergens', 'allergens_fr', 'arachidic-acid_100g', 'bicarbonate_100g'], axis = 1, inplace = True)

#Remplissage des colonnes identifiées comme ayant un taux de remplissage >35% et <80%
#liste des colonnes à remplir non complète juste pour montrer la commande python

liste_col_aremplir = ['saturated-fat_100g', 'sugars_100g', 'trans-fat_100g']
for i in liste_col_aremplir:
    txrempl_col=(len(df)-sum(df[i].isnull()))/len(df)
    print("Traitement de REMPLISSAGE des valeurs manquantes par la médiane car remplie à hauteur de {}%.".format(100*txrempl_col)) 
    moy_col=df.dropna(subset=[i])[i].median()
    df[i].fillna(value=moy_col,inplace=True)
    Ntxrempl_col=(len(df)-sum(df[i].isnull()))/len(df)        
    print("Médiane de la colonne : {}.".format(str(moy_col)))
    print("Taux de remplissage de {} est désormais de : {} ".format(i,str(Ntxrempl_col)))

print(len(df.columns))
Nbr_lignes=len(df.index)
print(Nbr_lignes)
```

```python
# 2.1.1 Deuxième option: utiliser un calcul de comparaison itératif pour couvrir toutes les colonnes

for i in df.columns:
    print(i) 
    #txrempl_col=len(df[i].dropna())/len(df) 
    txrempl_col=(len(df)-sum(df[i].isnull()))/len(df)
    print(txrempl_col)
    
    if txrempl_col < 0.35:
        print("Traitement de SUPPRESSION de la colonne car n'est remplie qu'à hauteur de {}%.".format(100*txrempl_col))
        del df[i]
        
    else:
        if txrempl_col < 0.80 and df.dtypes[i]==float:
            print("Traitement de REMPLISSAGE des valeurs manquantes par la médiane car remplie à hauteur de {}%.".format(100*txrempl_col))
            moy_col=df.dropna(subset=[i])[i].median()
            df[i].fillna(value=moy_col,inplace=True)
            Ntxrempl_col=(len(df)-sum(df[i].isnull()))/len(df)        
            print("mediane de la colonne : {}.".format(str(moy_col)))
            print("Taux de remplissage de {} est désormais de : {} ".format(i,str(Ntxrempl_col)))
        else:
            print("AUCUN Traitement sur cette colonne")
            
print(len(df.columns))
Nbr_lignes=len(df.index)
print(Nbr_lignes)            
```

```python
# 2.1.2 Traitement des lignes avec un taux de valeurs manquantes supérieur 20%

bad_index = []
for j in df.index:
    
    txNrempl_ligne=sum(df.loc[j].isnull())/len(df.columns)

    if txNrempl_ligne > 0.20:
        bad_index.append(j)
         
# Ne garder que les indices qui ne sont pas dans la liste bad_index
df = df[~df.index.isin(bad_index)] 
#df = df [ ~condition ]=== base ou condition n'est pas  verifie

#ordonner les index
df.reset_index(inplace=True, drop=True)


print(len(df.columns))
Nbr_lignes=len(df.index)
print(Nbr_lignes)
```

```python
# 2.2 DATA CLEANING - DONNEES ABERRANTES OU EXTRÊMES
```

```python
# Traitement des outliers avec le GRAPHE Box PLOT

# 2.2.1 Première option: traitement des outliers avec vizualisation graphique avec Box PLOT

#Box plot pour tous les paramètres numériques
index_colnum = []
for i in df.columns:
    #If colunm is numerical, visualise data and identify outliers with the graph
    if df.dtypes[i]==float:
        index_colnum.append(i)

for k in index_colnum:
    sns.boxplot(y=df[k])
    plt.xlabel(k)
    plt.ylabel("Quantité")
    plt.title('graphe sur {}'.format(k))           
    #plt.show()

#Box plot pour trois paramètres identifiés comme ayant des outliers durant le Data profiling
liste_outliers_suspectés = ['salt_100g', 'sugars_100g', 'energy_100g']

print("Box plot AVANT traitement des outliers")
for i in liste_outliers_suspectés:  
    sns.boxplot(y=df[i])
    plt.xlabel(i)
    plt.ylabel("Quantité")
    plt.title('graphe sur {}'.format(i))           
    plt.show()
    
for i in liste_outliers_suspectés:   
    print(i)
    Quantile1 = df[i].dropna().quantile(.25)
    Quantile3 = df[i].dropna().quantile(.75)
    print("Q1:{}".format(Quantile1))
    print("Q3:{}".format(Quantile3))
    Inter_Quantile = Quantile3 - Quantile1
    print("IQ: {}".format(Inter_Quantile))
       
    bad_index = []
    for j in df.index:
              
        if ((df[i][j].any()) < (Quantile1 - 3*Inter_Quantile)) or ((df[i][j].any()) > (Quantile3 + 3*Inter_Quantile)):
            print("Q1-3*IQ: {}".format(Quantile1 - 3*Inter_Quantile))
            print("Q3+3*IQ: {}".format(Quantile3 + 3*Inter_Quantile))
            print("{} est un outlier".format(df[i][j]))
            bad_index.append(j)
                               
df = df[~df.index.isin(bad_index)]     
df.reset_index(inplace=True, drop=True)

print("Box plot APRES traitement des outliers")
for i in liste_outliers_suspectés:  
    sns.boxplot(y=df[i])
    plt.xlabel(i)
    plt.ylabel("Quantité")
    plt.title('graphe sur {}'.format(i))           
    plt.show()
```
